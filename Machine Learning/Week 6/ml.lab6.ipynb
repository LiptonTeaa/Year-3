# Linear Regression

## The Normal Equation

```python
# generate a dataset to illustrate linear regression based on the normal equation (see lecture on linear regresion)

import numpy as np
X = 2 * np.random.rand(100, 1) # 100 rows and 1 column
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
plt.plot(X, y, "b.") # the b. just means blue markers with round shape
plt.xlabel("$x_1$", fontsize=18) # the dollar signs give it the formatting in the graph
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
save_fig("generated_data_plot")
plt.show()
```

The regression coefficient signifies the amount by which change in x must be multiplied to give the corresponding average change in y, or the amount y changes for a unit increase in x. In this way it represents the degree to which the line slopes upwards or downwards.
<br> For linear regression, input variables are multiplied by the coefficient and added together. 
<br> Coefficient of the input variable (θ1) plus the bias (θ0), the parameters, multiplied by the input vaiable/variables (linear equation).
<br> Linear regeression in vectorial form, dot product between the vector of the parameters and the vector of input features (an instance in your data set).

Instance: A single row of data is called an instance. It is an observation from the domain. <br>
Feature: A single column of data is called a feature.

```python
# calculate the coefficients of the regression used in the Normal Equation (see lecture on linear regression)

X_b = np.c_[np.ones((100, 1)), X]  # add x0 = 1 to each instance (adding 1 to each row basically, new column)
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# This basically helps us find the optimal values (parameters) of the regression coefficients that minimize the 
# mean squared error between predicted values and actual values.
```

 So, computing the best parameters for the model here which is theta_best, we're using a known formula to make this
 calculation, <br>
 matrix for the training set features > transpose it > multiply with original matrix > compute inverse of reuslting matrix > multiply again with the transpose of the original matrix > finally multiply with the outcome variable vector.
 <br> SVD is a more practical method of doing this, <br>
 a matrix X+ derived from the initial training set multiplied with the outcome variable. <br>
 X+ is the pesudoinverse of X (mateix of the training set features).
 

```python
print("Array type =", X.dtype, '\n')
print("Array type =", X_b.dtype, '\n')
```

```python
# display coefficients of linear regression (including intercept)
# parameter values are the coefficients of the features. Intercept is like our bias, 4.19888802 here.

theta_best
```

```python
# predict on vector X_new containing 2 data points

X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]  # add x0 = 1 to each instance; np.c_ concatenate arrays
# this added a second column of 1's to each row, making it a 2x2 matrix from a 2x1 matrix)
y_predict = X_new_b.dot(theta_best) # this corresponds to a linear combination of features and their coefficients
y_predict 
```

I guess that theta_best is basically the learned parameters of the model we calculated, like weights and biases 
having already been set the correct values through training on the previous areay X_b. It uses this to calculate the 
predicted values y_predict and transform the input features X_new_b. This helps us plot our best fit line through
the original data points fromt he first array.

```python
# plot the linear regression line passing through the 2 predicted points

plt.plot(X_new, y_predict, "r-") # best fit line with new data
plt.plot(X, y, "b.") # scatter points for original data 
plt.axis([0, 2, 0, 15])
plt.show()
```

```python

# Here is a better plot doing the same thing 

plt.plot(X_new, y_predict, "r-", linewidth=2, label="Predictions")
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([0, 2, 0, 15])
save_fig("linear_model_predictions_plot")
plt.show()
```

# Gradient Descent
## Batch Gradient Descent

```python
# This illustrates how the Batch Gradient Descent works (see lecture on linear regression) 

eta = 0.1  # learning rate
n_iterations = 1000
m = 100

theta = np.random.randn(2,1)  # random initialization

for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients
```

```python
# Display coefficients computed by the Batch Gradient Descent method

theta
```

```python
# Display prediction made with this method on new data

X_new_b.dot(theta)
```

```python

# Let us now see in plots how Batch Gradient Descent progressively converges depending on learning rate
# See lecture for details

theta_path_bgd = []

def plot_gradient_descent(theta, eta, theta_path=None):
    m = len(X_b)
    plt.plot(X, y, "b.")
    n_iterations = 1000
    for iteration in range(n_iterations):
        if iteration < 10:
            y_predict = X_new_b.dot(theta)
            style = "b-" if iteration > 0 else "r--"
            plt.plot(X_new, y_predict, style)
        gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
        theta = theta - eta * gradients
        if theta_path is not None:
            theta_path.append(theta)
    plt.xlabel("$x_1$", fontsize=18)
    plt.axis([0, 2, 0, 15])
    plt.title(r"$\eta = {}$".format(eta), fontsize=16)
```

```python
np.random.seed(42)
theta = np.random.randn(2,1)  # random initialization

plt.figure(figsize=(10,4))
plt.subplot(131); plot_gradient_descent(theta, eta=0.02)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.subplot(132); plot_gradient_descent(theta, eta=0.1, theta_path=theta_path_bgd)
plt.subplot(133); plot_gradient_descent(theta, eta=0.5)

save_fig("gradient_descent_plot")
plt.show()
```

## Stochastic Gradient Descent

```python

# Let us now see in plots how Stochastic Gradient Descent works, step by step
# This is updating the parameters (coefficients) step by step based on an instance at a time
# See lecture for details


theta_path_sgd = []
m = len(X_b)
np.random.seed(42)

n_epochs = 50
t0, t1 = 5, 50  # learning schedule hyperparameters

def learning_schedule(t):
    return t0 / (t + t1)

theta = np.random.randn(2,1)  # random initialization

for epoch in range(n_epochs):
    for i in range(m):
        if epoch == 0 and i < 20:                    # not shown in the book
            y_predict = X_new_b.dot(theta)           # not shown
            style = "b-" if i > 0 else "r--"         # not shown
            plt.plot(X_new, y_predict, style)        # not shown
        random_index = np.random.randint(m)
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
        theta_path_sgd.append(theta)                 # not shown

plt.plot(X, y, "b.")                                 # not shown
plt.xlabel("$x_1$", fontsize=18)                     # not shown
plt.ylabel("$y$", rotation=0, fontsize=18)           # not shown
plt.axis([0, 2, 0, 15])                              # not shown
save_fig("sgd_plot")                                 # not shown
plt.show()                                           # not shown
```

```python
# The model's coefficients

theta
```

```python
# All this computation can be done by using the SGDRegressor class in sklearn
# Fitting such a SGDRegressor model will comprise all the iterations illustrated above

from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor(max_iter=1000, tol=1e-3, penalty=None, eta0=0.1, random_state=42)
sgd_reg.fit(X, y.ravel())
```

## Mini-batch gradient descent

```python
# Mini-batch gradient descent is use small batches, and is somewhere in between the above 2 methods. 
# Let's illustrate it (see lecture for details)

theta_path_mgd = []

n_iterations = 50
minibatch_size = 20

np.random.seed(42)
theta = np.random.randn(2,1)  # random initialization

t0, t1 = 200, 1000
def learning_schedule(t):
    return t0 / (t + t1)

t = 0
for epoch in range(n_iterations):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(0, m, minibatch_size):
        t += 1
        xi = X_b_shuffled[i:i+minibatch_size]
        yi = y_shuffled[i:i+minibatch_size]
        gradients = 2/minibatch_size * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(t)
        theta = theta - eta * gradients
        theta_path_mgd.append(theta)
```

```python
# Display coefficients

theta
```

```python
# Let us now compare the 3 methods, from the point of view of how their computations converge

theta_path_bgd = np.array(theta_path_bgd)
theta_path_sgd = np.array(theta_path_sgd)
theta_path_mgd = np.array(theta_path_mgd)

plt.figure(figsize=(7,4))
plt.plot(theta_path_sgd[:, 0], theta_path_sgd[:, 1], "r-s", linewidth=1, label="Stochastic")
plt.plot(theta_path_mgd[:, 0], theta_path_mgd[:, 1], "g-+", linewidth=2, label="Mini-batch")
plt.plot(theta_path_bgd[:, 0], theta_path_bgd[:, 1], "b-o", linewidth=3, label="Batch")
plt.legend(loc="upper left", fontsize=16)
plt.xlabel(r"$\theta_0$", fontsize=20)
plt.ylabel(r"$\theta_1$   ", fontsize=20, rotation=0)
plt.axis([2.5, 4.5, 2.3, 3.9])
save_fig("gradient_descent_paths_plot")
plt.show()
```

# Polynomial Regression

```python
import numpy as np
import numpy.random as rnd

np.random.seed(42)
```

```python
# Generate a non-linear (quadratic, power 2) data with noise to ilustrate how a more powerful Polynomial regression works

m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)
```

```python
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([-3, 3, 0, 10])
save_fig("quadratic_data_plot")
plt.show()
```

```python
# Adding new featurs comprising such non-linear data to our dataset can automatically be done by using PolynomialFeatures function

from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
X[0]
```

```python
# In this case, for each feature value x, x^2 is added to that instance as a new feature value  
# Indeed (-0.75275929)^2 = 0.56664654

X_poly[0]
```

```python
# Now we can fit a linear regression to the new data. 
#  Its model will act as if it is computed by a polynimial (more powerful function that can fit the data more flexibly)

lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
# Let us represent data and predictions in a plot

X_new=np.linspace(-3, 3, 100).reshape(100, 1)
X_new_poly = poly_features.transform(X_new)
y_new = lin_reg.predict(X_new_poly)
plt.plot(X, y, "b.")
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([-3, 3, 0, 10])
save_fig("quadratic_predictions_plot")
plt.show()
```

```python

# Let us now compate 3 models of polynomial regression with degress of 1 (red line), 2 (blue curve), and 300 (green curve)
# We see how polynomial regression for degree 300 overfits data because too flexible;
# We see also how polynomial regression for degree 1 (linear regression) undefits data because not flexible enough; 
#  We see how polynomial regression for degree 2 captures well the pattern in the data.

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

for style, width, degree in (("g-", 1, 300), ("b--", 2, 2), ("r-+", 2, 1)):
    polybig_features = PolynomialFeatures(degree=degree, include_bias=False)
    std_scaler = StandardScaler()
    lin_reg = LinearRegression()
    polynomial_regression = Pipeline([
            ("poly_features", polybig_features),
            ("std_scaler", std_scaler),
            ("lin_reg", lin_reg),
        ])
    polynomial_regression.fit(X, y)
    y_newbig = polynomial_regression.predict(X_new)
    plt.plot(X_new, y_newbig, style, label=str(degree), linewidth=width)

plt.plot(X, y, "b.", linewidth=3)
plt.legend(loc="upper left")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([-3, 3, 0, 10])
save_fig("high_degree_polynomials_plot")
plt.show()
```

# Learning Curves

```python
# Define a funcion for plotting learning curves

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=10)
    train_errors, val_errors = [], []
    for m in range(1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))

    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="train")
    plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="val")
    plt.legend(loc="upper right", fontsize=14)   # not shown in the book
    plt.xlabel("Training set size", fontsize=14) # not shown
    plt.ylabel("RMSE", fontsize=14)              # not shown
```

```python
# We now illustrate how the linear regression underfits the data and that more data does not help improving error which remains high
# See explanations in lecture (and recommended chapter)

lin_reg = LinearRegression()
plot_learning_curves(lin_reg, X, y)
plt.axis([0, 80, 0, 3])                         # not shown in the book
save_fig("underfitting_learning_curves_plot")   # not shown
plt.show()                                      # not shown
```

```python

# We now illustrate how the polynmial regression with large degree = 10 overfits the data and that more data help improving error which decreases
# See explanations in lecture (and recommended chapter)


from sklearn.pipeline import Pipeline

polynomial_regression = Pipeline([
        ("poly_features", PolynomialFeatures(degree=10, include_bias=False)),
        ("lin_reg", LinearRegression()),
    ])

plot_learning_curves(polynomial_regression, X, y)
plt.axis([0, 80, 0, 3])           # not shown
save_fig("learning_curves_plot")  # not shown
plt.show()                        # not shown
```
